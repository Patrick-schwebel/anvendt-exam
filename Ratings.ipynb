```python

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Function to calculate similarity with stopwords
def calculate_similarity_with_stopwords(df1, df2):
    vectorizer = TfidfVectorizer(stop_words='english')
    combined_texts = [' '.join(df1['reviewText'].dropna()), ' '.join(df2['reviewText'].dropna())]
    vectors = vectorizer.fit_transform(combined_texts)
    similarity = cosine_similarity(vectors[0:1], vectors[1:])
    return np.mean(similarity) * 100

# Function to plot rating class similarity heatmap with stopwords
def plot_rating_class_similarity_heatmap_with_stopwords(ax, df, category):
    rating_classes = sorted(df['overall'].unique())
    similarity_matrix = np.zeros((len(rating_classes), len(rating_classes)))

    for i in range(len(rating_classes)):
        for j in range(len(rating_classes)):
            subset_df1 = df[df['overall'] == rating_classes[i]]
            subset_df2 = df[df['overall'] == rating_classes[j]]

            similarity_percentage = calculate_similarity_with_stopwords(subset_df1, subset_df2)
            similarity_matrix[i, j] = similarity_percentage

    sns.heatmap(similarity_matrix, annot=True, fmt=".1f", cmap="coolwarm", xticklabels=rating_classes, yticklabels=rating_classes, ax=ax)
    ax.set_title(f"Rating Class Similarity in {category} (with stopwords)")
    ax.set_xlabel('Rating Class')
    ax.set_ylabel('Rating Class')

# Combine all categories into one DataFrame
combined_df = pd.concat(category_data.values(), ignore_index=True)

# Plot the combined heatmap
plt.figure(figsize=(12, 8))
plot_rating_class_similarity_heatmap_with_stopwords(plt.gca(), combined_df, "Combined Categories")
plt.show()
```
